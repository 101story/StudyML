## Axes와 Subplot

여러개의 플롯을 배열 형태로 보여야하는 경우 <br>
각각의 플롯은 Axes <br>
http://matplotlib.org/api/axes_api.html#matplotlib.axes.Axes


subplot 명령을 사용 하기 위해 Axes 객체를 얻음 <br>
그리드(grid) 형태의 Axes 객체들을 생성<br>
Figure가 행렬(matrix)이고 Axes가 행렬의 원소

subplot(행, 열, 몇번쨰)

```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
# cos 코사인 함수 
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

ax1 = plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')
print(ax1)

ax2 = plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')
print(ax2)

plt.show()
```

## 바차트 
 bar 세로 barh 가로

http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.bar <br>
http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.barh

```python
y = [2, 3, 1]
x = np.arange(len(y))
xlabel = [u'a', u'b', u'c']
plt.bar(x, y)
plt.xticks(x, xlabel)
plt.show()
```

- 에러 바(error bar)를 추가 : xerr, yerr 이용 
- alpha는 투명도 :  0이면 완전 투명, 1이면 완전 불투명

```python
people = (u'A', u'B', u'C', u'D', u'E')
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

plt.barh(y_pos, performance, xerr=error, align='center', alpha=0.4)
plt.yticks(y_pos, people)
plt.xlabel(u'x label')
plt.show()
```

두개의 바 차트 한번에 그리기 

```python
n_groups = 5
means_men = (20, 35, 30, 35, 27)
std_men = (2, 3, 4, 1, 2)
means_women = (25, 32, 34, 20, 25)
std_women = (3, 5, 2, 3, 3)

fig, ax = plt.subplots()

index = np.arange(n_groups)
bar_width = 0.35

opacity = 0.4
error_config = {'ecolor': '0.3'}

rects1 = plt.bar(index, means_men, bar_width,
                 alpha=opacity,
                 color='b',
                 yerr=std_men,
                 error_kw=error_config,
                 label='Men',
                 align='edge')

rects2 = plt.bar(index + bar_width, means_women, bar_width,
                 alpha=opacity,
                 color='r',
                 yerr=std_women,
                 error_kw=error_config,
                 label='Women',
                 align='edge')

plt.xlabel('Group')
plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(index + bar_width, ('A', 'B', 'C', 'D', 'E'))
plt.legend()
plt.tight_layout()
plt.show()
```

## 컨투어 플롯
입력 변수가 x, y 두 개, 출력 변수가 z  하나인 경우 3차원<br>
등고선(contour) : contour 혹은 contourf 명령 사용 

http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.contour <br>
http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.contourf

```python
def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 -y ** 2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
XX, YY = np.meshgrid(x, y)
ZZ = f(XX, YY)
ZZ
```

```python
plt.contourf(XX, YY, ZZ, alpha=.75, cmap='jet')
plt.contour(XX, YY, ZZ, colors='black', linewidth=.5)
plt.show()
```

## 삼각 그리드 생성

Triangulation 클래스 <br>
x, y는 일련의 점들의 x좌표와 y좌표를 나타내는 1차원 벡터 <br>
triangles 어떤 삼각형이 있으며 각 삼각형이 어떤 점들로 이루어져있는가

```python
x = np.asarray([0, 1, 2])
y = np.asarray([0, np.sqrt(3), 0])
triangles = [[0, 1, 2]]
triang = mtri.Triangulation(x, y, triangles)
plt.triplot(triang, 'ko-')
plt.xlim(-0.1, 2.1)
plt.ylim(-0.1, 1.8)
plt.show()
```

```python
x = np.asarray([0, 1, 2, 3, 4, 2])
y = np.asarray([0, np.sqrt(3), 0, np.sqrt(3), 0, 2*np.sqrt(3)])
triangles = [[0, 1, 2], [2, 3, 4], [1, 2, 3], [1, 3, 5]]
triang = mtri.Triangulation(x, y, triangles)
plt.triplot(triang, 'ko-')
plt.xlim(-0.1, 4.1)
plt.ylim(-0.1, 3.7)
plt.show()
```

## factorplot

색상(hue)과 행(row) 등을 동시에 사용하여 3 개 이상의 카테고리 값의 분포 변화


```python
sns.factorplot(x="age", y="embark_town", hue="sex", row="class", data=titanic[titanic.embark_town.notnull()],
               size=2, aspect=3.5, kind="violin", split=True)
plt.show()
```

여러 종류의 차트를 겹치기 

```python
sns.boxplot(x="tip", y="day", data=tips, whis=np.inf)
sns.stripplot(x="tip", y="day", data=tips, jitter=True, color="0.4")
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=tips, inner=None)
sns.swarmplot(x="day", y="total_bill", data=tips, color="0.9")
plt.show()
```
