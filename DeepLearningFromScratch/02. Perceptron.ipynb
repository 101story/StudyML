# 2.1 퍼셉트론 이란?

퍼셉트론 : 다수의 신호(흐름)를 입력받아 하나의 신호를 출력
    
x = 뉴런 = 노드 <br> 
w = 가중치 <br>
임계값 : xw 총 합이 정해진 한계를 넘을때만 1을 출력<br>


# 2.3 퍼셉트론 구하기

```python
# and 
def AND(x1, x2):
    w1, w2, thera = 0.5, 0.5, 0.7
    tmp = x1*w1 + x2*w2
    if tmp <= thera:
        return 0
    elif tmp > thera:
        return 1
```

```python
print(AND(0,0))
print(AND(0,1))
print(AND(1,0))
print(AND(1,1))
```

* 가중치, 편향

```python
x = np.array([0,1])
w = np.array([0.5,0.5])
b = -0.7
print(w*x)
print(np.sum(w*x))
print(np.sum(w*x)+b)
```

```python
def AND(x1, x2):
    x = np.array([x1,x2])
    w = np.array([0.5,0.5])
    b = -0.7
    tmp = np.sum(w*x)+b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
# NAND
def NAND(x1, x2):
    x = np.array([x1,x2])
    w = np.array([-0.5,-0.5])
    b = 0.7
    tmp = np.sum(w*x)+b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
# OR
def OR(x1, x2):
    x = np.array([x1,x2])
    w = np.array([0.5,0.5])
    b = -0.2
    tmp = np.sum(w*x)+b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
print(AND(0,0))
print(AND(0,1))
print(AND(1,0))
print(AND(1,1))
print('----------')
print(NAND(0,0))
print(NAND(0,1))
print(NAND(1,0))
print(NAND(1,1))
print('----------')
print(OR(0,0))
print(OR(0,1))
print(OR(1,0))
print(OR(1,1))
```

# 2.5 다층 퍼셉트론

```python
def XOR(x1,x2):
    s1 = NAND(x1,x2)
    s2 = OR(x1,x2)
    y = AND(s1,s2)
    return y
```

```python
print(XOR(0,0))
print(XOR(0,1))
print(XOR(1,0))
print(XOR(1,1))
```
